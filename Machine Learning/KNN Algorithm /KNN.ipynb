```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix

%matplotlib inline
```

```python
df = pd.read_csv('/Users/aneruthmohanasundaram/Documents/GitHub/Machine_Learning/Machine Learning/KNN Algorithm /classified_data.csv')
df.head()
```

```python
scaler = StandardScaler()
```

```python
# Considering only feature class
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
# Converting the feature class into array format (scaled value)
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1))

# To check the scaled value
scaled_features
```

```python
# Use the scaled_feature varible to recreate the feature data frame
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])

# To check the df_feat
# This is the standardised verion of the data  
df_feat
```

```python
# To split the data into two methods
# Test data set and Train data set 

X = df_feat
y = df['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
# Initialise the KNN classifier and assigning the meighbors as 1
lm = KNeighborsClassifier(n_neighbors=1)

# Fitting the model to the train data 
lm.fit(X_train,y_train)
```

```python
# Predicting the test data 
pre = lm.predict(X_test)

# To display the predicted data 
pre
```

```python
# To print the confusion matrix 
print('Confusion Matrix:' + '\n''\n',confusion_matrix(y_test,pre))

print('\n')

# To print clasification report 
print('Classification Report')
print('\n')
print(classification_report(y_test,pre))
```

```python
# Using elbow method to choose the correct K value.  
rate = []

for i in range(1,40):
    lm1 = KNeighborsClassifier(n_neighbors=i)
    lm1.fit(X_train,y_train)
    predict = lm1.predict(X_test)
    rate.append(np.mean(predict != y_test)) # to check the prediction with the test data

# TO visualise the prediction
plt.figure(figsize=(10,6))
plt.plot(range(1,40),rate,color='blue',linestyle='--',marker='o',markerfacecolor='red',markersize=10) # to print the predictions for range of 1 to 40.
plt.title('Error Rate vs K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# After seeing the above chart choose the higher K value
knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)

# To print the confusion matrix 
print('Confusion Matrix:' + '\n''\n',confusion_matrix(y_test,pred) )
# print('\n')
# print(confusion_matrix(y_test,pre))

print('\n')

# To print clasification report 
print('Classification Report')
print('\n')
print(classification_report(y_test,pred))
```
