```python
import seaborn as sns 
import matplotlib.pyplot as plt 

from sklearn.cluster import KMeans
# make an artificial data using scikit
from sklearn.datasets import make_blobs

%matplotlib inline
```

```python
# initialise the blob 
data = make_blobs(n_samples=200, n_features=2, centers=None,cluster_std=1.8,random_state=101)
```

```python
# ploting the blobs 
plt.figure(figsize=(10,6))
# plotting the data with the all columns and data with first columns 
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

```python
# calling the kmeans classifier 
km = KMeans(n_clusters=2)

# fiting above plotted data 
km.fit(data[0])

# fetching the centers of the clusters 
print('The central point for the clusters is:'+ '\n',km.cluster_centers_)

print('\n')

# to get the labels 
print('Labels for the clusters is:'+'\n',km.labels_)
```

```python
#  To compare orgina ldata with predicted data 
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=km.labels_,cmap='rainbow')

ax2.set_title('Orginal')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
