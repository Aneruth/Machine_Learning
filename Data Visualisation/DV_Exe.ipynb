```python
import numpy as np
import pandas as pd
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.DataFrame(np.random.randn(30,3),columns=['a','b','c'])
print(df.info())
df.head()
```

```python
# To create a scatter plot which plots a points b vs a in red color with 12X3 Figsize
df.plot.scatter(x='a',y='b',figsize=(12,3),c='red')
```

```python
df.hist(column='a',bins=10)
```

```python
# or we can do like this 
df['a'].plot.hist()
```

```python
# Create a box plot to compare a and b columns 
df[['a','b']].plot.box()
```

```python
# Create a area plot and show rows upto 30
df.plot.area(alpha=0.1,stacked=False)
# or we can use df.ix[0:30].plot.area(alpha=0.1,stacked=False)
```

```python
# Create a KDE plot for c Column 
df['c'].plot.kde()
```

```python
# line style dashed KDE Plot 
df['c'].plot.kde(lw=5,ls='--',color='red')
```
