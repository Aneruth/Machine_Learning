```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
# create a numpy array 
x = np.linspace(1,2,10)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Fucntional Method 
plt.plot(x,y,'y--')
plt.xlabel('Abi')
plt.ylabel('Ane')
plt.title('Sample Matplotlib')
```

```python
plt.subplot(1,2,1)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('plot1')
plt.plot(x,y,'b--')

plt.subplot(1,2,2)
plt.title('plot2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
ane = np.arange(10).reshape(5,2)
ane1 = ane **2
plt.plot(ane1,ane,'r--')
```

```python
# Object Oriented Method 
```

```python
fig = plt.figure()
axes = fig.add_axes([0.56,0.98,1,0.88])
axes.set_xlabel('X-Axis')
axes.set_ylabel('Y-Axis')
axes.set_title('Object Oriented Table')
axes.plot(ane1)
```

```python
plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig
```

```python
fig1,axes1 = plt.subplots(nrows=1,ncols=2)
axes1[0].plot(x**10,y,'r')
axes1[0].set_title('Plot 1')
axes1[0].set_xlabel('X Axis')
axes1[0].set_ylabel('Y Axis')

axes1[1].plot(y**3,x,'b')
axes1[1].set_title('Plot 2')
axes1[1].set_xlabel('X Axis')
axes1[1].set_ylabel('Y Axis')

plt.tight_layout()
```

```python
axes1
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,4))

axes[0].set_xlabel('X Axis')
axes[0].set_ylabel('Y Axis')
axes[0].set_title('Sample Chart')
axes[0].plot(x**2)


axes[1].set_xlabel('X Axis')
axes[1].set_ylabel('Y Axis')
axes[1].set_title('Sample Chart')
axes[1].plot(y**5,x)

fig.savefig('Matplot Output.png',dpi=850)
plt.tight_layout()
```

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])

ax.plot(x,y**3,label='To the power 4')
ax.plot(x,y**4,label='To the power 5')

ax.legend(loc='best')
```

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])

ax.plot(x,y*12,label='Plot 1')
ax.plot(x,y*10,label='Plot 2')

ax.legend(loc=(0.1,0.8))
```

```python
x1 = [1.5,2.4,3.2,4.6,5.1]
y1 = [6.4,7.2,8.3,9.1,10.0]

abi = plt.figure()

adi = abi.add_axes([0,1,1,1.5])
adi.set_xlabel('X - Axis')
adi.set_ylabel('Y - Axis')
adi.set_title('Plot')
adi.plot(x1,y1,color='red',lw=1,linestyle='-')
adi.set_xlim([1.5,5])
adi.set_ylim([6.5,9])
```

```python
# Default Use of code 
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a new figure and add the axes to the figure canvas at [0,0,1,1] as ax. Plot (x,y)
liv = plt.figure()

ax = liv.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Use subplots top to create a two rows and a single column graph
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,2))
axes[0].plot(x,y,'b--')
axes[1].plot(y,z,'r-')
```
